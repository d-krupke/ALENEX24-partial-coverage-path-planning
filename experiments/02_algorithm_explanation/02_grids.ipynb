```python
from pcpptc import PolygonInstance
from pcpptc.instance_converter import RotatingRegularSquare, RotatingRegularHexagonal
from pcpptc.instance_converter.dmsh import DmshGrid

from pcpptc.plot import setup_plot, plot_polygon_instance
import matplotlib.pyplot as plt
```

```python
from pcpptc.plot.intermediate import plot_graph

pi = PolygonInstance.from_json(file_path="./01b_instance.json")
pi.expensive_areas = []
ax = setup_plot()
plot_polygon_instance(ax, pi, plot_valuable_areas=False)
plt.show()
```

```python
figsize = (3.5, 3.5)
```

```python
t = RotatingRegularSquare(full_coverage=True, point_based=False)
gi1 = t(pi)
```

```python
ax = setup_plot(figsize=figsize)
plot_polygon_instance(ax, pi)
plot_graph(ax, gi1.graph, marker=".", vertex_color="black", edge_color="black")
plt.tight_layout()
plt.savefig("02a_square_grid.pdf")
plt.show()
```

```python
t = RotatingRegularHexagonal(full_coverage=True, point_based=False)
gi2 = t(pi)
```

```python
ax = setup_plot(figsize=figsize)
plot_polygon_instance(ax, pi)
plot_graph(ax, gi2.graph, marker=".", vertex_color="black", edge_color="black")
plt.tight_layout()
plt.savefig("02b_hex_grid.pdf")
plt.show()
```
