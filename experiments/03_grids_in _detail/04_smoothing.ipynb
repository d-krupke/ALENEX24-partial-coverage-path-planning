```python
import os


def out_path(filename):
    path = "./04_out/"
    os.makedirs(path, exist_ok=True)
    return os.path.join(path, filename)
```

```python
from pcpptc import PolygonInstance
from pcpptc.instance_converter import RotatingRegularSquare, RotatingRegularHexagonal
from pcpptc.instance_converter.dmsh import DmshGrid

from pcpptc.plot import setup_plot, plot_polygon_instance
import matplotlib.pyplot as plt
```

```python
from pcpptc.plot.intermediate import plot_graph

pi = PolygonInstance.from_json(file_path="./01b_instance.json")
ax = setup_plot()
plot_polygon_instance(ax, pi, plot_valuable_areas=False)
plt.show()
```

```python
figsize = (3.5, 3.5)
```

```python
from pcpptc.instance_converter.dmsh import GmshGrid

t = GmshGrid(
    full_coverage=True,
    point_based=False,
    buffer=0.1,
    simplification=0.1,
    alg=1,
    opt_method="",
)
gi5 = t(pi)
```

```python
ax = setup_plot(figsize=figsize)
# ax = setup_plot()
plot_polygon_instance(ax, pi)
plot_graph(ax, gi5.graph, marker=".", vertex_color="black", edge_color="black")
plt.tight_layout()
plt.savefig(out_path("without_smoothing.pdf"))
plt.show()
```

```python
from pcpptc.instance_converter.dmsh import GmshGrid

t = GmshGrid(
    full_coverage=True, point_based=False, buffer=0.1, simplification=0.1, alg=1
)
gi5 = t(pi)
```

```python
ax = setup_plot(figsize=figsize)
# ax = setup_plot()
plot_polygon_instance(ax, pi)
plot_graph(ax, gi5.graph, marker=".", vertex_color="black", edge_color="black")
plt.tight_layout()
plt.savefig(out_path("with_smoothing.pdf"))
plt.show()
```
